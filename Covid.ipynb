## Che cos è il COVID-19?

__CoViD-19__ (*Corona Virus Desease 19*) è il nome della malattia provocata dal virus **SARS-CoV-2**, ossia una malattia infettiva respiratoria. Il virus appartiene alla famiglia dei coronavirus.

I primi casi sono stati registrati nella città di Wuhan nella provincia dell Hubei in Cina. Da lì si è rapidamente diffusa in tutto il mondo, costringendo L'OMS a dichiararla come pandemia.

Il Virus si diffonde da Uomo a Uomo attraverso uno stretto contatto tra individui, in particolare tossendo e starnutendo su qualcun altro che si trova entro un raggio di circa 1-2 metri dal infetto. È oltresì possibile infettarsi toccando superfici ove sia presente il virus.

Attualmente le strategie migliori per combattere l' epidemia sono
1. Distanziamento sociale: osserversare una distanza interpersonale di 1-2 metri
2. Utilizzo di Dispositivi di protezione individuale, quali guanti e mascherine PF3
3. Lockdown completo delle regioni/nazioni colpite

Ad oggi i casi registrati nel mondo sono: 3.300.724 mentre 1.037.926 persone si sono guarite e sono morte 233.693 persone per l'infezione.

## La situazione Italiana

In Italia il primo caso di Covid-19 è stato registrato il 30 gennaio 2020, in una coppia di turisti cinesi. Il 31 gennaio il governo italiano ha dichiarato stato d'emergenza. Il primo caso di trasmissione secondario sul suolo nazionale è stato registrato il 18 febbraio dello stesso anno.

Al progredire del infezione, il governo ha emanato vari provvedimenti e decreti legge per far fronte. Al 30 aprile sono in vigore le seguenti misure:
1. Divieto alle persone fisiche di trasferirsi o spostarsi con mezzi di trasporto pubblici o privati un comune diverso da quello in cui si trovano, salvo che per comprovate esigenze lavorative, di assoluta urgenza ovvero per motivi di salute.
2. Chiusura delle attività produttive non essenziali o strategiche. Restano aperti alimentari, farmacie, negozi di generi di prima necessità e i servizi essenziali.

## Analisi Dati 

Il Dipartimento dell protezione civile ha pubblicato e mantine aggiornata una repository su [GitHub](https://github.com/pcm-dpc/COVID-19), in cui giornalmente vengono inseriti i nuovi dati riguardanti l'epidemia in corso.

Lo scopo di questo report è analizzare i dati e mostrare a grafico i risutlati.

### Set up

Si è provveduto a fare un fork della suddetta repository. La repository è organizzata in cartelle gerchiche, le informazioni sono organizzate in formato CSV e JSON. Ai fini di questa analisi si sono considerati i dati nazionali contenuti nel file dati-json/dpc-covid19-ita-andamento-nazionale-latest.json e sono così organizzati:

| Nome campo                  | Descrizione                       | Description                            | Formato                       | Esempio             |
|-----------------------------|-----------------------------------|----------------------------------------|-------------------------------|---------------------|
| **data**                        | Data dell'informazione            | Date of notification                   | YYYY-MM-DD HH:MM:SS (ISO 8601) Ora italiana | 2020-03-05 12:15:45 |
| **stato**                       | Stato di riferimento              | Country of reference                   | XYZ (ISO 3166-1 alpha-3)      | ITA                 |
| **ricoverati_con_sintomi**      | Ricoverati con sintomi            | Hospitalised patients with symptoms    | Numero                        | 3                   |
| **terapia_intensiva**           | Ricoverati in terapia intensiva   | Intensive Care                         | Numero                        | 3                   |
| **totale_ospedalizzati**        | Totale ospedalizzati              | Total hospitalised patients            | Numero                        | 3                   |
| **isolamento_domiciliare**      | Persone in isolamento domiciliare | Home confinement                       | Numero                        | 3                   |
| **totale_positivi** | Totale attualmente positivi (ospedalizzati + isolamento domiciliare)      | Total amount of current positive cases (Hospitalised patients + Home confinement)  | Numero                        | 3                   |
| **variazione_totale_positivi**  | Variazione del totale positivi (totale_positivi giorno corrente - totale_positivi giorno precedente)       | News amount of current positive cases (totale_positivi current day - totale_positivi previous day)  | Numero                        | 3                   |
| **nuovi_positivi**  | Nuovi attualmente positivi (totale_casi giorno corrente - totale_casi giorno precedente)       | News amount of current positive cases (totale_casi current day - totale_casi previous day)  | Numero                        | 3                   |
| **dimessi_guariti**             | Persone dimesse guarite           | Recovered                              | Numero                        | 3                   |
| **deceduti**                    | Persone decedute                  | Death                                  | Numero                        | 3                   |
| **totale_casi**                 | Totale casi positivi              | Total amount of positive cases         | Numero                        | 3                   |
| **tamponi**                     | Totale tamponi                    | Tests performed                        | Numero                        | 3                   |
| **casi_testati**                     | Totale dei soggetti sottoposti al test                    | Total number of people tested                        | Numero                        | 3                   |
| **note_it**                     | Note in lingua italiana (separate da ;)                   | Notes in italian language (separated by ;)                       | Testo                        | pd-IT-000                   |
| **note_en**                     | Note in lingua inglese (separate da ;)                    | Notes in english language (separated by ;)                       | Testo                        | pd-EN-000                   |



Per effettuare l' analisi si è utilizzato Python (3.8.1) con il paccheto SciPy. Di Seguito il codice utilizzato e il risultato dello stesso.

```python
import json
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
import datetime as dt
import numpy as np

with open("COVID-19/dati-json/dpc-covid19-ita-andamento-nazionale.json", "r") as jsonData:
    data = jsonData.read()
    
# leggi json in una struttura python
andamenti = json.loads(data)

# le colonne d' interesse
columns = ['data', 'ricoverati_con_sintomi', 'terapia_intensiva', 'totale_ospedalizzati', 'isolamento_domiciliare', 'totale_positivi', 'variazione_totale_positivi', 'nuovi_positivi', 'dimessi_guariti', 'deceduti', 'totale_casi', 'tamponi', 'casi_testati']

dati = {col: [] for col in columns}

# trasformazione in ina lista di dizionari, così da poter usare un dataframe pandas
for andamento in andamenti:
    for col in columns:
        dati[col].append(andamento[col])
        
df = pd.DataFrame(dati)
df["data"] = pd.to_datetime(df["data"], infer_datetime_format=True)
df["data"] = df["data"].dt.strftime("%d/%m")

df1 = df[['data', 'nuovi_positivi']]
df1['tamponi_giornalieri'] = df['tamponi'].diff()
df1['pos%'] = df1['nuovi_positivi']/df1['tamponi_giornalieri']
```

```python
# Tamponi univoci
print("Tamponi univoci effettuati: ", df['casi_testati'].iloc[-1])

# Percentuali di casi totali positivi rispetto al numero univoco di tamponi effettuati
perc_pos_tamp_unit = df['totale_casi'].iloc[-1]/df['casi_testati'].iloc[-1] * 100

# medie
medie = df1.mean(axis=0, numeric_only=True)

print("Percentuale di positivi totali rispetto al numero univico di tamponi: ", perc_pos_tamp_unit)
print("\nMedie")

print("Numero medio di nuovi positivi: ", round(df1['nuovi_positivi'].mean()))
print("Numero medio di tamponi gionralieri: ", round(df1['tamponi_giornalieri'].mean()))
print("Percentuale media di nuovi positivi giornalieri: ", df1['pos%'].mean() * 100)
```

```python
# La Curva dei casi totali

df.plot(kind="line", x="data", y="totale_casi", grid=True, figsize=(30,10), rot=10, marker='o')
```

```python
# Curva dei nuovi positivi

df.plot(kind="line", x="data", y="totale_positivi", grid=True, figsize=(30,10), rot=10, marker='o')
```

```python
# confronto delle due curve

df.plot(kind="line", x="data", y="totale_casi", grid=True, figsize=(30,10), rot=10)
plt.plot(df["data"], df["totale_positivi"])
```

```python
# Curva nuovi positivi su totale casi confermati in scala logaritmica

plt.figure(figsize=(30,20))
plt.plot(df['totale_casi'], df['nuovi_positivi'], 'b')
plt.xscale('log')
plt.yscale('log')
plt.grid()
```

```python
# Grafico per i tamponi giornalieri effettuati e positivi trovati

df1.plot(kind="bar", x="data", y=['nuovi_positivi', 'tamponi_giornalieri'], grid=True, figsize=(30,10), rot=10)
```

```python
# Percentuale di positivi gionalieri su tamponi giornalieri effettuati

df1.plot(kind="bar", x="data", y='pos%', grid=True, figsize=(30,10), rot=10)
```
